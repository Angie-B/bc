# Display relative changes to surface temperature and precipitation by country.

```python
%matplotlib inline
from matplotlib import pyplot as plt
import sqlite3

QUERY = '''select A.year, (A.value + ?4)/(B.value + ?4)
from readings A join readings B on A.year=B.year and A.what=B.what
where A.what=?1 and A.country=?2 and B.country=?3;'''

def get_ratio(db_name, what, country_A, country_B, offset=0.0):
    '''Fetch ratio of values (by year) for two countries from database.'''
    connection = sqlite3.connect(db_name)
    cursor = connection.cursor()
    cursor.execute(QUERY, [what, country_A, country_B, offset])
    results = cursor.fetchall()
    years = []
    values = []
    for (y, v) in results:
        years.append(y)
        values.append(v)
    return years, values
```

```python
years, values = get_ratio('climate.db', 'tas', 'CAN', 'AUS', 273.15)
plt.plot(years, values)
plt.show()
```
